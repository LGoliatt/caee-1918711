````python
# Additional Function Example: Factorial Calculation
def factorial_iterative(n):
    """Calculate factorial of a number using iteration."""
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Testing the factorial functions
number = 5
print(f"\nFactorial of {number} (Iterative):", factorial_iterative(number))
````

## Introduction to Pandas

### What is Pandas?

Pandas is a powerful and flexible Python library for data manipulation and analysis.
It provides data structures and functions designed to make working with structured data easy and intuitive.
Built on top of NumPy, Pandas leverages its efficiency for numerical operations.
Essentially, Pandas brings the power of spreadsheets and relational databases to Python.

### Key Features

* Data Structures:
    * Series: One-dimensional labeled array, similar to a column in a spreadsheet or a NumPy array with labels.
    * DataFrame: Two-dimensional labeled data structure, like a table or spreadsheet, with columns of potentially different data types.
* Data Manipulation: Offers functions for filtering, sorting, grouping, merging, and reshaping data.

### Pandas in Action

* Pandas is used for:
    * Data cleaning and preprocessing.
    * Exploratory data analysis (EDA).
    * Data visualization.
    * Statistical analysis.
    * Building machine learning pipelines.

# Data Science Challenge: **Water Quality Monitoring in Uruguay**

## Dataset Overview

The dataset provided at the link ([Water Quality Monitoring](https://catalogodatos.gub.uy/dataset/monitoreo-de-calidad-en-los-cuerpos-de-agua/resource/ecb68681-9eb2-45b5-97af-dcf95f4e7a53)) contains information about water quality monitoring in various water bodies across Uruguay. It includes parameters such as:

- **Location**: Geographic details of the water body.
- **Date**: The date of the measurement.
- **Water Quality Parameters**: Metrics like pH, dissolved oxygen, turbidity, temperature, and other indicators of water quality.
- **Pollutants**: Levels of contaminants or pollutants (e.g., nitrates, phosphates, heavy metals).
- **Compliance Status**: Whether the water body meets environmental quality standards.

This dataset is ideal for exploring real-world environmental issues and applying data science techniques to analyze and address them.
