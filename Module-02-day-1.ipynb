## Accessing and Understanding Data from the Cat치logo de Datos Uruguay

**Objective:** To familiarize yourselves with open data sources and begin the process of understanding and interpreting real-world datasets.

**Task:**

1.  **Access the Cat치logo de Datos Uruguay:**
    * Navigate to the [Cat치logo de Datos Uruguay](https://catalogodatos.gub.uy).
2.  **Find Interesting Data:**
    * Explore the catalog and identify datasets that pique your interest. Consider environmental, social, or economic data.
3.  **Form Groups:**
    * Organize yourselves into groups of 2-4 people.
4.  **Download the Data:**
    * Download the selected dataset(s) in a suitable format (CSV, JSON, etc.).
5.  **Understand the Data:**
    * Examine the data structure, variables, and any accompanying documentation.
    * Consider:
        * What are the variables?
        * What are the units of measurement?
        * What is the time period covered by the data?
6.  **Tell a Story:**
    * **Can you tell a story from the data?**
    * Discuss within your group:
        * What potential insights can be gained from the data?
        * What questions can be answered?
        * Can you create a narrative based on the data?

**Deliverable:**

* A brief summary of the dataset(s) you explored.
* A description of your initial understanding of the data.
* A short discussion of the "story" you believe the data tells.

**Note:** This task is designed to be exploratory. Focus on understanding the data and brainstorming potential analyses.
