## Raw Data vs. Processed Data
---
**Raw Data:**  
- [Original source of data](https://valhalla1.openstreetmap.de/route?json={"locations":[{"lat":-34.919208756647066,"lon":-56.16655111312866},{"lat":-34.91674110030358,"lon":-56.1629730463028}],"costing":"pedestrian","directions_options":{"units":"kilometers"}})
- Usually difficult to use in data analysis.  
- Requires processing before analysis.  

**Processed Data:**  
- [Ready for analysis](https://valhalla.openstreetmap.de/directions?profile=pedestrian&wps=-56.1629730463028%2C-34.91674110030358%2C-56.16655111312866%2C-34.919208756647066)  
- Processing includes merging, subsetting, transformations.  
- Rules must be defined for processing.  
- All steps must be documented.  
- Examples: Normalization, outlier handling, removal of irrelevant values.

## Introduction to Data Storytelling
---
### What is Data Storytelling?
Data storytelling is the practice of using data to tell compelling and insightful stories. It involves:
- **Data Analysis:** Extracting meaningful insights from data.
- **Narrative:** Crafting a coherent and engaging story around the data.
- **Visualization:** Using charts, graphs, and other visual tools to illustrate the data.

### Importance
> Data storytelling helps communicate complex information in an accessible and impactful way, making it easier for audiences to understand and act upon the insights.

### Connection to a paper
The paper [Habitat changes and changing predatory habits in North American fossil canids](https://www.nature.com/articles/ncomms8976) is an excellent example of data storytelling. It uses:
- **Data Analysis:** Examining elbow-joint shapes to infer predatory behaviors.
- **Narrative:** Describing how climatic and environmental changes influenced the evolution of canids.
- **Visualization:** Presenting findings that correlate habitat changes with the emergence of specialized predatory techniques.
