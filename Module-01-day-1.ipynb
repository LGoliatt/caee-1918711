---

## Introduction to Data Storytelling
---
### What is Data Storytelling?
Data storytelling is the practice of using data to tell compelling and insightful stories. It involves:
- **Data Analysis:** Extracting meaningful insights from data.
- **Narrative:** Crafting a coherent and engaging story around the data.
- **Visualization:** Using charts, graphs, and other visual tools to illustrate the data.

### Importance
> Data storytelling helps communicate complex information in an accessible and impactful way, making it easier for audiences to understand and act upon the insights.

## Paper overview
---
### Introduction
This study explores how climatic and environmental changes over the past 37 million years have influenced the evolution of predatory behaviors in North American canids.

## Habitat Changes and Changing Predatory Habits in North American Fossil Canids

### Habitat Changes
- Analysis of elbow-joint shapes to infer predatory behaviors.
- Specialized species first appeared with the spread of open habitats in the late Oligocene.
- Late Miocene saw the emergence of elbow-joint morphologies indicative of modern pounce-pursuit predators.
- Shift in plant communities from C3 to C4 grasses.

### Predatory Habits
- Pursuit canids first appeared during the Pleistocene.
- Evolution of **predatory** techniques correlated with habitat changes.

## Key Findings
- Climate change and its impact on vegetation and habitat structure are critical for the emergence of ecological landscape.
- Environmental transformations profoundly influence the evolution of predatory behaviors in canids.

- Demonstrates the importance of habitat changes in shaping the direction of lineage evolution.
- Highlights the role of environmental factors in the development of specialized predatory techniques.
